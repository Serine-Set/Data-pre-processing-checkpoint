When dealing with missing values, there are several techniques that can be used to replace them. These include:
    -Replacing missing values with the mode: Mode is the most frequently occurring value in a column. 
        This method is useful when the data is categorical.
# Replace missing values with mode : data['column_name'].fillna(data['column_name'].mode()[0], inplace=True)

    -Replacing missing values with the mean: Mean is the average value of a column.
        This method is useful when the data is numerical.
# Replace missing values with mean : data['column_name'].fillna(data['column_name'].mean(), inplace=True)

    Replacing missing values with the median: Median is the middle value of a column when it is sorted in ascending order. 
        This method is useful when the data is numerical and has outliers.
# Replace missing values with mean data['column_name'].fillna(data['column_name'].mean(), inplace=True)
